## 03 | Principal Component Plots

This notebook contains code for visualizing the Principal Components obtained from the Principal Component Analysis on offspring genotype data.

**** AgAdapt Project ****

- Import required libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

- Load Principal Component data into a Pandas DataFrame.

```python
pc_df = pd.read_hdf("../Data/C__Processed_Data/Offspring_Genotype_Data/03__Offspring_Principal_Components/Offspring_Principal_Components.h5", "Offspring")
pc_df
```

- Create a composite plot of the first 40 principal components of each individual.

```python
figure = plt.figure(figsize = (20, 20))

for i in range(1, 21):

    # Add subplot per new graph.
    pc_plot = figure.add_subplot(5, 4, i)

    # Convert target principal components of main data to arrays.
    pc_x = pc_df.iloc[:, i * 2 - 2].to_numpy()
    pc_y = pc_df.iloc[:, i * 2 - 1].to_numpy()

    # Create axis labels and grid.
    pc_plot.set_xlabel("PC " + str(i * 2 - 1), fontsize = 12)
    pc_plot.set_ylabel("PC " + str(i * 2), fontsize = 12)
    pc_plot.set_facecolor("#F5F5F5")
    pc_plot.grid(True, linestyle = ":")

    # Remove axis numbering.
    pc_plot.axes.get_xaxis().set_ticks([])
    pc_plot.axes.get_yaxis().set_ticks([])

    # Plot main data.
    for x, y in zip(pc_x, pc_y):
        pc_plot.plot(x, y, ".", color = "#0055FF")
```

- Save the plot to a file.

```python
figure.savefig("../Results/Plots/Offspring_Principal_Component_Analysis/Principal_Component_Plot.png", facecolor = "w", dpi = 500)
```

- Import subsample names retrieved from the visual inspection of the popVAE plot.
- This allows comparison between the two methods.

```python
subsample_df = pd.read_csv("../Data/C__Processed_Data/Supplemental_Data/Reference_Subsample_Names.txt", names = ["Sample"])
subsample_df
```

- For each individual in the subsample data, retrieve the respective principal components.

```python
subsample_df = pc_df[pc_df.index.isin(subsample_df["Sample"])]
subsample_df
```

- Recreate the plot with the individuals from the subsample highlighted.
- While the principal components **do** cluster these similar individuals, without the explicit highlighting this cannot be visually distinguished.

```python
figure = plt.figure(figsize = (20, 20))

for i in range(1, 21):

    # Add subplot per new graph.
    subsample_plot = figure.add_subplot(5, 4, i)

    # Convert target principal components of main data to arrays.
    pc_x = pc_df.iloc[:, i * 2 - 2].to_numpy()
    pc_y = pc_df.iloc[:, i * 2 - 1].to_numpy()

    # Convert target principal components of subsample data to arrays.
    ss_x = subsample_df.iloc[:, i * 2 - 2].to_numpy()
    ss_y = subsample_df.iloc[:, i * 2 - 1].to_numpy()

    # Create axis labels and grid.
    subsample_plot.set_xlabel("PC " + str(i * 2 - 1), fontsize = 12)
    subsample_plot.set_ylabel("PC " + str(i * 2), fontsize = 12)
    subsample_plot.set_facecolor("#F5F5F5")
    subsample_plot.grid(True, linestyle = ":")

    # Remove axis numbering.
    subsample_plot.axes.get_xaxis().set_ticks([])
    subsample_plot.axes.get_yaxis().set_ticks([])

    # Plot main data.
    for x, y in zip(pc_x, pc_y):
        subsample_plot.plot(x, y, ".", color = "#0055FF")

    # Plot subsample data.
    for x, y in zip(ss_x, ss_y):
        subsample_plot.plot(x, y, ".", color = "#FF0055")
```

- Save the plot to a file.

```python
figure.savefig("../Results/Plots/Offspring_Principal_Component_Analysis/Principal_Component_Plot_Highlighted_Subsample.png", facecolor = "w", dpi = 500)
```
